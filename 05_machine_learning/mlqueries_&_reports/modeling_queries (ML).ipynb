-- Training with BOOSTED TREE REGRESSOR

```python
%%bigquery
CREATE OR REPLACE MODEL `yellow-taxi-trips-2026.ml_dataset.boosted_tree_regressor`
  OPTIONS (model_type="BOOSTED_TREE_REGRESSOR", enable_global_explain=TRUE, input_label_cols=["total_amount"])
AS 
SELECT * FROM `yellow-taxi-trips-2026.ml_dataset.preprocessed_train_data`;
```

```python
%%bigquery
SELECT COUNT(*) FROM `yellow-taxi-trips-2026.ml_dataset.preprocessed_test_data`;
```

-- Evaluate the trained model with the test data

```python
%%bigquery
SELECT * FROM 
ML.EVALUATE(MODEL `yellow-taxi-trips-2026.ml_dataset.boosted_tree_regressor`, 
(SELECT * FROM `yellow-taxi-trips-2026.ml_dataset.preprocessed_test_data`));
```

-- Example for making predictions from the model

```python
%%bigquery
SELECT * FROM
ML.PREDICT (MODEL `yellow-taxi-trips-2026.ml_dataset.boosted_tree_regressor`, 
(SELECT * FROM `yellow-taxi-trips-2026.ml_dataset.preprocessed_test_data`));
```

-- Query the model's global explanations

```python
%%bigquery
SELECT * FROM ML.GLOBAL_EXPLAIN(MODEL `yellow-taxi-trips-2026.ml_dataset.boosted_tree_regressor`);
```
